```python
from guided_diffusion import image_datasets
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
```

```python

bs = 8
ds = image_datasets.load_data(
      data_dir="/mnt/data_rome/ffhq/images1024x1024/",
      batch_size=bs,
      image_size=512, # input scale of images
      random_resize=True, # resize between 1024-512
      random_crop=True, # random crop the rest of the way to 512
      out_size=128, # Size to pass to network (will be degraded and further random cropped)
      scale_factor=2, # downsampling factor for bsrgan
     )
```

```python
for batch in ds:
    plt.figure(figsize=(20,20))
    g = make_grid(batch[1]["low_res"], nrow=4)
    plt.imshow(g.permute(1,2,0))
    plt.axis("off")
    plt.show()
    break
```
